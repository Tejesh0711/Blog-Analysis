```python
#convert all values to numeric
def dummyEncode(df):
        columnsToEncode = list(df.select_dtypes(include=['category','object']))
        le = LabelEncoder()
        for feature in columnsToEncode:
            try:
                if feature!='Deliquency':
                    df[feature] = le.fit_transform(df[feature])
                else:
                    print('Hello')
                    df[feature] = df[feature]
            except:
                print('Error encoding '+feature)
        return df
```

```python
#Neural Network
#neural network
scaler = StandardScaler()
# Fit only to the training data
scaler.fit(xtrain)
xtrain_nn = scaler.transform(xtrain)
xtest_nn = scaler.transform(xtest)

neural_network_class=MLPClassifier()
neural_network_class.fit(xtrain_nn,ytrain)
predictions = neural_network_class.predict(xtest_nn)
y_pred=predictions

#calculate ROC curve
preds = neural_network_class.predict_proba(xtest_nn)[:,1]
calculate_roc_curve(ytest, preds,2)
 #calculate Confusion Matrix
calculate_confusion_matrix(ytest, y_pred)

print(accuracy_score(ytest, y_pred))
```
