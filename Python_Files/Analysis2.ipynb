```python
#importing required packages
import webhoseio
import requests
import numpy as np
import pandas as pd
import json
import os
import time
import datetime
```

```python
#getting current working directory
current_directory=os.getcwd()
root_directory=os.path.abspath(os.path.join(current_directory, os.pardir)) #To get the parent directory of current directory
#parent_directory=os.path.abspath(os.path.join(root_directory, os.pardir))
print(root_directory)
```

```python
path=root_directory+"/Data/Processed Data/"
```

```python
processeddf=pd.read_csv(path+"PocessedData.csv",encoding = "ISO-8859-1")
```

```python
#Getting country wise daily blog count 
x1=processeddf.groupby(["published_day","country"])["uuid"].count().reset_index(name="Blog count").sort_values(("Blog count"),ascending=False).head(20)
```

```python
#Getting country wise daily reply count
x2=processeddf.groupby(["published_day","country"])["replies_count"].sum().reset_index(name="Reply count").sort_values(("Reply count"),ascending=False).head(20)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure()
ax1 = fig.add_subplot(211)
ax2 = fig.add_subplot(212)

g1 = sns.factorplot(x="published_day", y="Blog count", hue="country", data=x1, ax=ax1)
g2 = sns.factorplot(x="published_day", y="Reply count", hue="country", data=x2, ax=ax2)

#g1.savefig('output.png')

plt.close(g1.fig)
plt.close(g2.fig)
plt.show()

#From this graph we can conclude that people in US contribute to blogs on a daily basis whereas that is not the case with people from other countries.
```
